```python
# Get the data 
conect_map=pd.read_csv("D:/ROSALIND_problems/Mongo_db/cocaine.csv")
conect_map = conect_map.iloc[: , 1:]
conect_map.head()
```

```python
conect_map["_id"] = conect_map["_id"].str.title()
conect_map["_id"] = conect_map["_id"].str.strip()
county_df["NAME"] = county_df["NAME"].str.title()
county_df["NAME"] = county_df["NAME"].str.strip()
conect_map.head()
# merge county_df with connect_map by name/county
res_w_states = county_df.merge(conect_map, left_on="NAME", right_on="_id", how="inner")
res_w_states.head()

res_w_states['value'].loc[res_w_states['value'].idxmin()]
```

```python
# Plotting 
fig, gax = plt.subplots(figsize = (10,5))
state_df[state_df['NAME'] == 'Connecticut'].plot(ax = gax, edgecolor='black',color='white')
res_w_states.plot(
    ax=gax, edgecolor='black', column='value', legend=True, cmap='RdBu_r',
    vmin=3, vmax=374.0

)
plt.rcParams["figure.figsize"] = (10,10.5)
gax.annotate('Death by fentanyl consumption',xy=(0.76, 0.06),  xycoords='figure fraction')

plt.axis('off')

plt.show()

```
